```python
!pip install bs4 # for install libraries
!pip install requests 

from bs4 import BeautifulSoup  # for import libraries
import requests

page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special') #to get requests to webpage

soup=BeautifulSoup(page.content)  #to get page content

titles=[]
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)  #scraping all titles of restaurents
    
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text) #scraping all location of restaurents
    
price=[]    
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text.split('|')[0]) #scraping all price of restaurents
    
images=[]
for i in soup.find_all('img',class_="no-img"):
    images.append(i.get('data-src'))
    
import pandas as pd # import pandas from python libraries
df=pd.DataFrame({'Titles':titles,'Location':location,'Price':price,'Images_url':images})
df






```
